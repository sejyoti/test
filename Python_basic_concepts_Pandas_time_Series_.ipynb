```python
#time series analysis

#3Explanation:

#Time series analysis in Pandas refers to analyzing data that is indexed by dates or timestamps.

#Some important aspects:

#- Pandas has built-in capabilities for working with time series data in a Series or DataFrame indexed by DatetimeIndex.

#- We can generate a time index using date_range() and easily create time series data.

#- Pandas has methods like .shift(), .re.rolsample(), ling() etc that are designed specifically for time series.
#These make lagging, downsampling, smoothing operations easy.

#- Time series specific plot types like autocorrelation_plot, lag_plot etc are available for visual analysis.

#So in summary, Pandas provides a lot of convenient tools for handling, manipulating and visualizing time series data in Python.

#The date/time indexes and time-aware methods make time series analysis much easier.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create time series data
dates = pd.date_range('01-01-2019', periods=36, freq='M')
data = [10, 13, 21, 28, 31, 29, 30, 32, 23, 19,
        15, 13, 16, 19, 25, 29, 24, 19, 15, 17,
        20, 26, 31, 35, 31, 30, 28, 26, 25, 24,
        26, 28, 26, 27, 32, 36]
ts = pd.Series(data, index=dates)

# Resample to yearly frequency
ts_year = ts.resample('A').mean()
print(ts_year)

# Rolling statistics
rol_mean = ts.rolling(window=12).mean()

# Plot original and resampled
ax = ts.plot(label='Monthly')
ts_year.plot(ax=ax, label='Yearly')
rol_mean.plot(ax=ax, label='Rolling 12 Months Average')
ax.legend()

plt.show()

# The output `Freq: A-DEC, dtype: float64` refers to the frequency and dtype attributes of the resampled DataFrame/Series.

#Specifically:

#- `A-DEC`: This is the frequency. `A` stands for annual, and `DEC` means the year ends in December.
#So this is a yearly frequency where each period represents the year ending in December.

#- `dtype: float64`: This shows the datatype of the values as 64-bit (double precision) floats.

#So in summary, by doing a resample to yearly frequency, pandas has generated a Series with annual data (ending in Dec), stored as float64 values.

#The key things the frequency (A-DEC) tells us are:

#- Resampling period is annual
#- Each value is the mean over a year period

#Understanding the frequency string is important to know what each data point represents when working with time series data in pandas.
```
